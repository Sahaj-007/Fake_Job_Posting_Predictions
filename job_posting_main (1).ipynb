```python
#Fill NULL values in string columns

string_cols = df.select_dtypes(include='object').columns  #object=text/string colns    #.columns--return only coln names

for col in string_cols:                             #Loop one by one through all text columns
    df[col] = df[col].fillna("Not Provided")        #Replacing NaN values

#  Fill Null values in numeric columns
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns   #int64=whole no's     #float64=decimal no's

for col in numeric_cols:
    df[col] = df[col].fillna(df[col].median())

print("Null values fixed successfully!")
print(df.isnull().sum())
```

TEXT CLEANING

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')
```

```python
#define a texxt processing fun
def clean_text(text):
  text=re.sub(r'[^\w\s]', '', text)
  text=text.lower()
  text=word_tokenize(text)
  text=[word for word in text if word not in stopwords.words('english')]
  return ' '.join(text)
```

```python
#Apply Preprocessing
df['clean_description'] = df['description'].apply(clean_text)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Combine useful text columns
df['text_data'] = df['title'] + " " + df['description'] + " " + df['requirements']
# generally fake clues exist in this 3

# Convert text to TF-IDF features
tfidf = TfidfVectorizer(max_features=5000)

X = tfidf.fit_transform(df['text_data']).toarray()

# Target column
y = df['fraudulent']
```

TRAIN-TEST SPLIT

```python
# Splitting the data for Fake Job Posting Project
# Taking input features (x)
x = np.array(df[['telecommuting', 'desc_length']])

# Taking target output (y) which we have to predict
y = np.array(df['fraudulent'])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report

x_train,x_test,y_train,y_test=train_test_split(x,y, test_size=0.20, random_state=42)

```

TRAINING A MODEL

1.Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

lr_model=LogisticRegression()
lr_model.fit(x_train, y_train)
```

```python
#Logistic Regression
lr_preds=lr_model.predict(x_test)
print("Logistic Regression")
print("Accuracy:", accuracy_score(y_test,lr_preds))
```

2.Naive Bayes

```python
from sklearn.naive_bayes import MultinomialNB

nb_model=MultinomialNB()
nb_model.fit(x_train,y_train)
```

```python
#Naive Bayes
nb_preds=nb_model.predict(x_test)
print("\nNaive Bayes")
print("Acurracy:", accuracy_score(y_test,nb_preds))
```

3.SVM

```python
from sklearn.svm import SVC

svm_model=SVC()
svm_model.fit(x_train,y_train)
```

```python
#Support Vector Machine
svm_preds=svm_model.predict(x_test)
print("\nSupport vector machine")
print("Acurracy:", accuracy_score(y_test,svm_preds))
```

4.Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier

rf_model=RandomForestClassifier()
rf_model.fit(x_train,y_train)
```

```python
rf_preds=rf_model.predict(x_test)
print("\nRandom forest")
print("Acurracy:",accuracy_score(y_test,rf_preds))
```

CONFUSION MATRIX---A confusion matrix shows what types of mistakes the model makes.

0=Real job


1=Fake job

```python
y_preds = rf_model.predict(x_test)

# Now use it in evaluation
from sklearn.metrics import confusion_matrix, accuracy_score

print("Accuracy:", accuracy_score(y_test, y_preds))
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_preds))
```

MODEL EVALUATION MATRIX--model evaluation metrics that measure how well the model predicts fraudulent (1) vs real (0) jobs.

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score
from sklearn.metrics import f1_score, confusion_matrix, classification_report

print("Accuracy :", accuracy_score(y_test, y_preds))   #measure overall correct predictions.
print("Precision:", precision_score(y_test, y_preds))  #Out of jobs predicted fake, how many were actually fake
print("Recall   :", recall_score(y_test, y_preds))     #Out of actual fake jobs, how many did model catch
print("F1 Score :", f1_score(y_test, y_preds))         #Balance of precision + recall.

print("\nConfusion Matrix")                            #Use to inspect false negatives
print(confusion_matrix(y_test, y_preds))


print("\nClassification Report")
print(classification_report(y_test, y_preds))

#Since fake job datasets are imbalanced, we used Recall and F1-score along with Accuracy to properly evaluate the classifier
```
